```python
import pandas as pd

df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
df.columns
```

```python
import prince
df2=df.drop('id',1)

yVals=df2['stroke']
df2=df2.drop('stroke',1)

df2['bmi'].fillna((df2['bmi'].mean()), inplace=True)

```

```python
mca = prince.MCA(
...     n_components=2,
...     n_iter=10000,
...     copy=True,
...     check_input=True,
...     engine='auto',
...     random_state=42
... )
mca_heart=mca.fit(df2)
toPlot=mca.transform(df2)
```

```python
ax = mca.plot_coordinates(
...     X=df2,
...     ax=None,
...     figsize=(6, 6),
...     show_row_points=True,
...     row_points_size=10,
...     show_row_labels=False,
...     show_column_points=True,
...     column_points_size=30,
...     show_column_labels=False,
...     legend_n_cols=1
... )
```

```python
import matplotlib.pyplot as plt

plt.scatter(toPlot[0],toPlot[1],c=yVals, alpha=0.1)
```
